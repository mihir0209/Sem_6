```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Task 1: Load and inspect the dataset.
df = pd.read_excel('sales_data.csv.xlsx')
print(df.head())
```

```python
#Task 2: Visualize the sales distribution across product categories using a bar plot.
plt.figure(figsize=(10,5))
sns.barplot(x='Category', y='Sales', color='Orange',data=df, estimator=sum, errorbar=None)
plt.title('Sales Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
#Task 3: Visualize the relationship between Sales and Profit using a scatter plot.
plt.figure(figsize=(8,5))
sns.scatterplot(x='Sales', y='Profit', data=df, alpha=0.6)
plt.title('Sales (X-axis) vs Profit (Y-axis)')
plt.show()
```

```python
#Task 4: Create a histogram to display the distribution of Quantity sold.
plt.figure(figsize=(8,5))
sns.histplot(df['Quantity'], bins=20, kde=True)
plt.title('Quantity Sold')
plt.show()
```

```python
#Task 5: Generate a box plot to compare Discounts given in each Region.
plt.figure(figsize=(8,5))
sns.boxplot(x='Region', y='Discount', data=df)
plt.title('Regional Discounts')
plt.xticks(rotation=45)
plt.show()
```

```python
# Task 6: Create a heatmap showing the correlation between different numerical variables.
plt.figure(figsize=(8,5))
corr_matrix = df.select_dtypes(include=['number']).corr()
sns.heatmap(corr_matrix, cmap='BuPu', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
