```python
import pandas as pd
import numpy as np
```

```python
#Task 1: Create a new column Revenue by multiplying Price and QuantitySold.
sales['Revenue_New'] = sales.apply(lambda row: row['Price'] * row['QuantitySold'], axis=1)
sales.head()
```

```python
#Task 2: Detect and remove outliers in the Price column using the IQR Method.
price_column = 'Price'
q1 = np.percentile(sales[price_column], 25)
q3 = np.percentile(sales[price_column], 75)
iqr= q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
```

```python
price_outliers = sales.loc[(sales[price_column] < lower_bound) | (sales[price_column] > upper_bound), price_column]
print(f"Outliers in {price_column}:\n{price_outliers}\n")
```

```python
sales_filtered = sales[(sales[price_column] >= lower_bound) & (sales[price_column] <= upper_bound)]

sales_filtered.head()
```

```python
sales_filtered.shape
```

```python
#Task 3: Normalize the Price and Revenue columns using Min-Max scaling.
from sklearn.preprocessing import MinMaxScaler

target_columns = ['Price', 'Revenue_New']
scaler = MinMaxScaler()
for col in target_columns:
    sales_filtered.loc[:, col] = scaler.fit_transform(sales_filtered[[col]])

sales_filtered.head()
```
